```python
import datetime as dt 
import requests 
import pandas as pd 
```

```python
BASE_URL = "http://api.openweathermap.org/data/2.5/weather?"
API_KEY = open("weatherapi.txt","r").read()
```

```python
data = pd.DataFrame(columns=["city","country","main_description","description","temperature","temperature_feels_like","temp_max","temp_min","pressure", "humidity", "wind_speed", "sea_level","sunrise", "sunset"])
```

```python
def cityname (city, printed):
    global response
    CITY = city
    url = BASE_URL + "appid=" + API_KEY + "&q=" + CITY
    response = requests.get(url).json()
    if response == {'cod': '404', 'message': 'city not found'}:
        return print(f"Town {city} not found")
    elif response != {'cod': '404', 'message': 'city not found'}:
        main_description = response["weather"][0]["main"]
        description = response["weather"][0]["description"]
        temperature = response["main"]["temp"] - 273.15
        temperature_feels_like = response["main"]["feels_like"] - 273.15
        temp_min = response["main"]["temp_min"] - 273.15
        temp_max = response["main"]["temp_max"] - 273.15
        pressure = response["main"]["pressure"]
        humidity = response["main"]["humidity"]
        wind_speed = response["wind"]["speed"]
        sea_level = response["wind"]["speed"]
        country = response["sys"]["country"]
        sunrise = dt.datetime.fromtimestamp(response["sys"]["sunrise"])
        sunset = dt.datetime.fromtimestamp(response["sys"]["sunset"])
        data.loc[len(data)] = [city,country,main_description, description,temperature,temperature_feels_like,temp_max,temp_min,pressure,humidity, wind_speed, sea_level, sunrise, sunset]
    if printed == "yes":
        print(f"City is {CITY}, {country}, with description - {description}")
        print(f"Current temperature: {temperature:.2f}, and it feels like {temperature_feels_like:.2f}")
        print(f"Today's max temperature: {temp_max:.2f}, and low: {temp_min:.2f}")
        print(f"Pressure: {pressure:.2f} and humidity: {humidity:.2f} with windspeed of {wind_speed:.2f} m/s")
        print(f"Sunrise at: {sunrise} and sunset at {sunset}")
    else:
        return
```

```python
for i in ["Bratislava","London","Berlin","Spisska Nova Ves","Nove Zamky","Rio de Janeiro","Brasil","Lima"]:
    cityname(i,"no")
```
